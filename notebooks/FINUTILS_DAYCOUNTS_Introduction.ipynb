# UNDERSTANDING DAY COUNTS

## Calculating year fractions according to different conventions

```python
from financepy.finutils import *
```

## Accrued Interest for a Bond

```python
numDays = 1
settlementDate = FinDate(15, 2, 2019)
frequencyType = FinFrequencyTypes.SEMI_ANNUAL

for dayCountType in FinDayCountTypes:

    pcd = FinDate(2019, 1, 1)
    ncd = pcd.addTenor("6M")
    dayCount = FinDayCount(dayCountType)

    alpha = dayCount.yearFrac(pcd, settlementDate, ncd, frequencyType)
    print("%45s YearFrac: %9.8f Days: %d  Period: %d" % (dayCount, alpha[0], alpha[1], alpha[2]))
```

## Year Fraction for Swaps

This should just take two dates.

```python
numDays = 1
settlementDate = FinDate(15, 2, 2019)
frequencyType = FinFrequencyTypes.SEMI_ANNUAL

for dayCountType in FinDayCountTypes:

    pcd = FinDate(2019, 1, 1)
    ncd = pcd.addTenor("3M")

    if dayCountType != FinDayCountTypes.ACT_ACT_ICMA:
        dayCount = FinDayCount(dayCountType)
        alpha = dayCount.yearFrac(pcd, ncd)
        print("%45s YearFrac: %9.8f Days: %d  Period: %d" % (dayCount, alpha[0], alpha[1], alpha[2]))
```
