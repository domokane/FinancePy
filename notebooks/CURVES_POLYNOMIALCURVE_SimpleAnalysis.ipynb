#  Examination of the FinPolynomialCurve

The polynomial curve parametrises the zero rate curve 

```python
import numpy as np
```

```python
from financepy.market.curves import *
from financepy.finutils import *
```

```python
import matplotlib.pyplot as plt
```

```python
times = np.linspace(0.01, 10.0, 20)
curveDate = FinDate(2019, 2, 2)
coeffs = [0.04, 0.0002, -0.000010]
```

```python
polyCurve = FinDiscountCurvePoly(curveDate, coeffs)
```

```python
dates = curveDate.addYears(times)
```

```python
zeros = polyCurve.zeroRate(dates)
fwds = polyCurve.fwd(dates)    
```

```python
plt.figure(figsize=(8, 6))
plt.plot(times, zeros*100.0, label="Zeros")
plt.plot(times, fwds*100.0, label="Forwards")
plt.xlabel('Time (years)')
plt.ylabel('Zero Rate (%)')
plt.legend(loc='best')
```
